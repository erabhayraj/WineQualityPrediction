```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
ds=pd.read_csv("winequalityN.csv")
ds.head(20)
```

```python
ds.fixedacidity=ds.fixedacidity.fillna(ds.fixedacidity.median())

ds.volatileacidity=ds.volatileacidity.fillna(ds.volatileacidity.median())

ds.citricacid=ds.citricacid.fillna(ds.citricacid.median())

ds.residualsugar=ds.residualsugar.fillna(ds.residualsugar.median())

ds.chlorides=ds.chlorides.fillna(ds.chlorides.median())

ds.freesulfurdioxide=ds.freesulfurdioxide.fillna(ds.freesulfurdioxide.median())

ds.totalsulfurdioxide=ds.totalsulfurdioxide.fillna(ds.totalsulfurdioxide.median())

ds.density=ds.density.fillna(ds.density.median())

ds.pH=ds.pH.fillna(ds.pH.median())

ds.sulphates=ds.sulphates.fillna(ds.sulphates.median())

ds.alcohol=ds.alcohol.fillna(ds.alcohol.median())
```

```python
ls=["fixedacidity","volatileacidity","citricacid","residualsugar","chlorides","freesulfurdioxide","totalsulfurdioxide","density","pH","sulphates","alcohol"]
reg = linear_model.LinearRegression()
reg.fit(ds[ls],ds.quality)
```

```python
plt.xlabel("Alcohol",fontsize="20")
plt.ylabel("Quality",fontsize="20")
plt.scatter(ds.alcohol,ds.quality,marker="*",color="red")
plt.plot(ds.alcohol,reg.predict(ds[ls]),color="blue")
```
